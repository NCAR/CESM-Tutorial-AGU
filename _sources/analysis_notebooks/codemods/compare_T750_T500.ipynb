### Import required modules

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from functions import *
import warnings
warnings.filterwarnings('ignore')
```

### Set up path to case h1 output

```python
file = "~/scratch/archive/b1850_T750/atm/hist/b1850_T750.cam.h1.0001-01-01-00000.nc"
```

### Open up the netcdf file

```python
dat = xr.open_dataset(file)
```

### Calculate the difference between them on day 1

```python
diff = (dat.T750 - dat.T500).isel(time=1)
```

### Specify contour interval and minimum and maximum contour values for the difference plot

```python
c_interval = 2 # contour interval
c_min = -30 # minimum contour
c_max = 30 # maximum contour
```

### Plot the plot

```python
fig = plt.figure(figsize=(16,16))

fig = plotmap(fig, diff, c_interval, c_min, c_max, 'T750 $-$ T500')
```
