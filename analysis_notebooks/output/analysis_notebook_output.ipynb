### Import required modules

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from functions import *
import warnings
warnings.filterwarnings('ignore')
```

### Set up path to case h0 output. 

```python
file = "~/scratch/archive/case01/atm/hist/case01.cam.h0.0001-01.nc"
```

### Open the netcdf file

```python
dat = xr.open_dataset(file)
```

### Specify the variable you want to plot (varplot) and the approximate pressure of the model level you'd like to plot (pplot)

```python
varplot='U'
pplot=500.
```

### Specify contour interval and minimum and maximum contour values

```python
c_interval = 2 # contour interval
c_min = -40 # minimum contour
c_max = 40 # maximum contuor
```

### Plot the plot

```python
fig = plt.figure(figsize=(16,16))

fig = plotmap(fig, dat[varplot].sel(lev=pplot, method='nearest').isel(time=0), c_interval, c_min, c_max, varplot+', ~'+str(pplot)+'hPa')
```
